```python
from sklearn.neural_network import MLPRegressor as nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler as ss
plt.rcParams['figure.figsize'] = [18, 12]
```

```python
files = ['../data/lucknow.csv',
         '../data/mirzapur.csv',
         '../data/noida.csv',
         '../data/unnao.csv',
         '../data/varanasi.csv']
demand = pd.read_csv('../data/up_consolidated.csv', parse_dates=['date_time']).set_index('date_time', drop=True)
demand.index = pd.to_datetime(demand.index)
hourly_avg = demand.groupby(pd.Grouper(freq='H')).mean()
```

```python
hourly_avg
```

```python
lk = pd.read_csv(files[0], parse_dates=['date']).set_index('date', drop=True)
lk.index = pd.to_datetime(lk.index).tz_localize(None)
```

```python
train = hourly_avg.join(lk, how='inner')
del train['precipitation_intensity']
del train['precipitation_probability']
del train['uv_index']
del train['icon']
train = train.fillna(0)
train = train.reset_index()
```

```python
train['index'] = (train['index'] - train['index'].min())
```

```python
train['days'] = train['index'].dt.days
train['hours'] = train['index'].dt.components['hours']
```

```python
del train['index']
```

```python
# train = train.loc[train['demand'] != 0]
```

```python
y_train = train.values[:,0]
x_train = train.values[:,1:]
```

```python
y_test = y_train[30000:]
x_test = x_train[30000:]
```

```python
y_train = y_train[:30000]
x_train = x_train[:30000]
```

```python
model = nn(hidden_layer_sizes=(5,7), max_iter=1000, learning_rate='adaptive',
           solver='adam')
scaler = ss()
```

```python
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
pd.DataFrame({'test': y_test, 'pred': y_pred}).plot()
```
