```python
%notebook inline
import pandas as pd
import xgboost as xgb
```

```python
bihar = pd.read_csv('../data/bihar.csv', na_values=' ').dropna().reset_index(drop=True).set_index('date')
```

```python
uttar = pd.read_csv('../data/uttar.csv', na_values=' ').dropna().reset_index(drop=True).set_index('date')
```

```python
bihar.index = pd.to_datetime(bihar.index)
```

```python
bihar.groupby(pd.Grouper(freq='M')).mean()['demand'].plot()
```

```python
# read in data
dtrain = xgb.DMatrix('../data/bihar_train.csv')
dtest = xgb.DMatrix('../data/bihar_test.csv')

# specify parameters via map
param = {'max_depth':2, 'eta':1, 'objective':'binary:logistic' }
num_round = 2
bst = xgb.train(param, dtrain, num_round)
```

```python
# make prediction
preds = bst.predict(dtest)
```
